```python
#Lendo as Imagens e Colocando em Tons de Cinza

img = cv2.imread('/content/imagem_trab.png')

#Lendo em BGR e RGB

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Posicionando lado a lado

img_final = cv2.hconcat((img, img2))

#Mostrando a Imagem

cv2_imshow(img_final)

#Passando para escalas de cinza

cinquenta_tons_de_cinza = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

#Mostrando a imagem novamente

cv2_imshow(cinquenta_tons_de_cinza)
```

```python
#Importando detector facial

face_detector = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

#Importando imagem para reconhecimento facial

fimg = cv2.imread('/content/imagem_trab.png')

#Pasando para cinza

fcimg = cv2.cvtColor(fimg, cv2.COLOR_BGR2GRAY)

#Utilizando Face Detector

dtr = face_detector.detectMultiScale(fcimg, scaleFactor = 1.1, minNeighbors = 3, minSize = (32, 32), maxSize = (100, 100))

#Desenhando os quadrados de detecção

for (x, y, w, z) in dtr:
  print(w, z)
  cv2.rectangle(img, (x, y), (x + w, y + z), (0, 255, 0), 2)

#Mostrando a imagem novamente

cv2_imshow(img)

```
